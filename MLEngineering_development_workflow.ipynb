### Data Science Project Template Directory

To provide a structure for the data science project life cycle and allow teams to work efficiently, several template structures have been proposed. Here, we will have a look at the template project proposed and maintained by the [Driven Data Team](https://www.drivendata.org/), and conveniently made available through cookiecutter (see the official documentation [here](https://drivendata.github.io/cookiecutter-data-science/)).

CookieCutter is a command line utility that assists you in creating projects from project templates. You can create a new project from the data science template by simply typing a command in your terminal (see data science project workflow tutorial). This will create the following structure:

image.png

You are not required to use all the files created, it is just a skeleton that is designed to help you work and collaborate on model development. Let’s have a look at the different main folders of this directory:

- **data**: is where you store your data. not only raw data, but also intermediate, processed data.
- **docs**: is where you document the code being developed and all technical aspects of the project
- **models**: is where you store snapshots of the models you produced
- **notebooks**: is where you put your jupyter notebooks! you can organize it in several subfolder for data exploration, modelling, etc…
- **reports**: is where you store reports to be shared outside the team
- **src**: is where you write your project’s package, ie an organized folder a file structure within which you write custom functions and classes that allow
